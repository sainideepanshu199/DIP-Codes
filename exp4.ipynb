```python
import cv2
import matplotlib.pyplot as plt
```

```python

image1 = cv2.imread('256x256(3).jpg')
image2 = cv2.imread('256x256(4).jpg')
image1.shape, image2.shape
```

```python

image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
```

```python
nand_result = cv2.bitwise_not(cv2.bitwise_and(image1_gray, image2_gray))
```

```python
nor_result = cv2.bitwise_not(cv2.bitwise_or(image1_gray, image2_gray))
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Image 1 (Color)')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Image 2 (Color)')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(nand_result, cmap='gray')
plt.title('NAND Result')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(image1_gray, cmap='gray')
plt.title('Image 1 (Grayscale)')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(image2_gray, cmap='gray')
plt.title('Image 2 (Grayscale)')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(nor_result, cmap='gray')
plt.title('NOR Result')
plt.axis('off')

plt.tight_layout()
plt.show()
```
