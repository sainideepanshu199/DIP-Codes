```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# image = cv2.imread('blob(4).webp', cv2.IMREAD_GRAYSCALE)
image = cv2.imread('blob(4).webp')
```

```python
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # X
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Y
```

```python
sobel_combined = cv2.addWeighted(cv2.convertScaleAbs(sobelx), 0.5, cv2.convertScaleAbs(sobely), 0.5, 0)
sharpened_image = cv2.addWeighted(image, 1, sobel_combined, 0.5, 0)
```

```python
plt.figure(figsize=(20, 12))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Operator')
plt.axis('off')
```
