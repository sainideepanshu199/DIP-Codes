```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_image = cv2.imread('920x517.webp')
resized_image = cv2.resize(original_image, (256, 256))
```

```python
mask_size = 100
masked_image = resized_image[128 - mask_size//2: 128 + mask_size//2, 128 - mask_size//2: 128 + mask_size//2]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1), plt.imshow(original_image[:, :, ::-1]), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(resized_image[:, :, ::-1]), plt.title('Resized Image')
plt.subplot(1, 3, 3), plt.imshow(masked_image[:, :, ::-1]), plt.title('Masked Image')
plt.show()
```
