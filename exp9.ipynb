```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def intensity_slice(image, low, high):
    if len(image.shape) > 2:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image

    mask = cv2.inRange(gray_image, low, high)
    enhanced_image = cv2.bitwise_and(image, image, mask=mask)

    return enhanced_image
```

```python
image = cv2.imread('blob(7).jpg')
low_intensity = 170
high_intensity = 425

enhanced_image = intensity_slice(image, low_intensity, high_intensity)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original Image
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[0].axis('off')

# Enhanced Image
axs[1].imshow(cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB))
axs[1].set_title('Enhanced Image')
axs[1].axis('off')

plt.show()
```
