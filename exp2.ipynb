```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('256x256(5).png', cv2.IMREAD_GRAYSCALE)
```

```python
min_intensity = np.min(image)
max_intensity = np.max(image)
stretched_image = cv2.convertScaleAbs(image, alpha=255.0/(max_intensity-min_intensity), beta=-min_intensity*(255.0/(max_intensity-min_intensity)))
```

```python
equalized_image = cv2.equalizeHist(image)
```

```python
hist_original = cv2.calcHist([image], [0], None, [256], [0,256])
hist_stretched = cv2.calcHist([stretched_image], [0], None, [256], [0,256])
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0,256])
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(stretched_image, cv2.COLOR_BGR2RGB))
plt.title('Contrast Stretched Image')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('Histogram Equalized Image')
plt.axis('off')

plt.subplot(2, 1, 2)
plt.plot(hist_original, color='black')
plt.plot(hist_stretched, color='red')
plt.plot(hist_equalized, color='blue')
plt.title('Histograms')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend(['Original', 'Stretched', 'Equalized'])

plt.tight_layout()
plt.show()
```
